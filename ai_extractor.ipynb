```python
import re

# Load notes
with open("meeting_notes.txt", "r") as f:
    notes = f.read()

print("RAW NOTES:\n")
print(notes)

# extarctor
tasks = []
lines = notes.split("\n")

for line in lines:
    if "waiting" in line or "needs" in line or "follow-up" in line or "confirm" in line:
        tasks.append(line.strip("- ").strip())

print("\nEXTRACTED ACTION ITEMS:\n")
for t in tasks:
    print("-", t)

# formatted output
structured = []
for t in tasks:
    owner = "Unknown"
    if "HR" in t:
        owner = "HR"
    elif "Finance" in t:
        owner = "Finance"
    elif "IT" in t:
        owner = "IT"

    structured.append({"task": t, "owner": owner})

print("\nSTRUCTURED OUTPUT:\n")
for item in structured:
    print(item)
```
